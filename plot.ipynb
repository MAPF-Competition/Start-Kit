```python
maps = ["Warehouse","Sortation","Random-5","Random-4","Random-3","Random-2","Random-1","Brc202d","Paris-2","Paris-1"]

new_errands = [66219,62708,2666,1573,1790,1063,956,6496,16489,5950]
new_agents = [10000,10000,800,600,400,200,100,6500,3000,1500]
new_time = [5000,5000,2000,1000,1000,500,500,5000,3500,1500]
new_dense = [26,18,98,73,49,24,12,15,6.4,3.2]
new_avg_dist = [272,238,27,27,27,27,27,452,202,202]

old_errands = [44782,57299,1245,1551,1936,1945,1921,5165,10733,8349]
old_agents = [5000,8000,800,600,400,200,100,3000,2000,1000]
old_time = [5000,5000,1000,1000,1000,1000,1000,5000,2500,2500]
old_dense = [13,15,98,73,49,24,12,6.9,4.2,2.1]
old_avg_dist = [272,238,27,27,27,27,27,452,202,202]

avg_manhattan_dist = [269.06,235.95,21.25,21.25,21.25,21.25,21.25,300.6,165.86,165.86]
max_manhattan_dist = [640, 640, 64, 64, 64, 64, 64, 1011,512,512]



```

```python
new_throughput = []
for i in range(0,len(new_errands)):
    new_throughput.append(new_errands[i]/new_time[i])

old_throughput = []
for i in range(0,len(old_errands)):
    old_throughput.append(old_errands[i]/old_time[i])
print(new_throughput)
print(old_throughput)
```

```python
new_throughput_per_agent = []
for i in range(0,len(new_errands)):
    new_throughput_per_agent.append(new_throughput[i]/new_agents[i])

old_throughput_per_agent = []
for i in range(0,len(old_errands)):
    old_throughput_per_agent.append(old_throughput[i]/old_agents[i])

print(new_throughput_per_agent)
print(old_throughput_per_agent)
```

```python
new_time_per_agent_task = []
for i in range(0,len(new_errands)):
    new_time_per_agent_task.append(1/new_throughput_per_agent[i])

old_time_per_agent_task = []
for i in range(0,len(old_errands)):
    old_time_per_agent_task.append(1/old_throughput_per_agent[i])

print(new_time_per_agent_task)
print(old_time_per_agent_task)
```

```python
new_avg_interaction = []
for i in range(0,len(new_errands)):
    new_avg_interaction.append(new_dense[i] * new_avg_dist[i])

old_avg_interaction = []
for i in range(0,len(old_errands)):
    old_avg_interaction.append(old_dense[i] * old_avg_dist[i])

print(new_avg_interaction)
print(old_avg_interaction)
```

```python

new_composite_difficulty = []
for i in range(0,len(new_errands)):
    new_composite_difficulty.append(new_avg_interaction[i])

old_composite_difficulty = []
for i in range(0,len(old_errands)):
    old_composite_difficulty.append(old_avg_interaction[i])
```

```python
new_normal_throughput = []
for i in range(0,len(new_errands)):
    new_normal_throughput.append(new_errands[i]/new_time[i] *(new_avg_dist[i]**2))

old_normal_throughput = []
for i in range(0,len(old_errands)):
    old_normal_throughput.append(old_errands[i]/old_time[i] * (old_avg_dist[i]**2))

print(new_normal_throughput)
print(old_normal_throughput)
```

```python
plt.scatter(new_agents,new_dense, color='green', marker="x")
# plt.scatter(old_agents,old_dense, color='blue', marker="+")
plt.xlabel("Num. of agents")
plt.ylabel("Density")
plt.show()

plt.scatter(new_dense,new_throughput, color='green', marker="x")
plt.scatter(old_dense,old_throughput, color='blue', marker="+")
plt.xlabel("Density")
plt.ylabel("Throughput")
plt.show()

plt.scatter(new_dense,new_normal_throughput, color='green', marker="x")
plt.scatter(old_dense,old_normal_throughput, color='blue', marker="+")
plt.xlabel("Density")
plt.ylabel("Normalized Throughput")
plt.show()


fig, ax = plt.subplots()
ax.scatter(new_avg_interaction,new_time_per_agent_task, color='green', marker="x")
ax.scatter(old_avg_interaction,old_time_per_agent_task, color='blue', marker="+")
# for i, txt in enumerate(maps):
#     ax.annotate(txt, (new_avg_interaction[i],new_time_per_agent_task[i]))
#     ax.annotate(txt, (old_avg_interaction[i],old_time_per_agent_task[i]))

plt.yscale('log')
plt.xscale('log')
plt.xlabel("Average Interaction")
plt.ylabel("Timesteps per agent per task")
plt.show()

fig, ax = plt.subplots()
ax.scatter(new_composite_difficulty,new_time_per_agent_task, color='green', marker="x")
ax.scatter(old_composite_difficulty,old_time_per_agent_task, color='blue', marker="+")
for i, txt in enumerate(maps):
    ax.annotate(txt, (new_composite_difficulty[i],new_time_per_agent_task[i]))
    ax.annotate(txt, (old_composite_difficulty[i],old_time_per_agent_task[i]))

plt.yscale('log')
plt.xscale('log')
plt.xlabel("Average Interaction Adjusted")
plt.ylabel("Time for an agent finish a task")
plt.show()


plt.scatter(new_avg_dist,new_throughput_per_agent, color='green', marker="x")
plt.scatter(old_avg_dist,old_throughput_per_agent, color='blue', marker="+")
plt.xlabel("Average distance")
plt.ylabel("Throughput per agent")
plt.show()


```
